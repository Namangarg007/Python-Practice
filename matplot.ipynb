# Matplotlib

* most of the work in matplot library is done using the modeule pyplot with contains all the functions required

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Plot function

* The plot() function is used to draw points (markers) in a diagram.

* By default, the plot() function draws a line from point to point.

* The function takes parameters for specifying points in the diagram.

* Parameter 1 is an array containing the points on the x-axis.

* Parameter 2 is an array containing the points on the y-axis.



```python
point_1 = np.array([5, 10])
point_2 = np.array([5, 10]) # first parameter in array represents x-axis and second parameter represents y-axis
# the points on axis are (5, 5) and (10, 10)
# we created two 1-D array to represent them in the graph using matplotlib
```

```python
point_1
```

```python
plt.plot(point_1, point_2) # in this point_1 will act as a point on the graph and point_2 will act as another point
plt.show()# this is used to show/display the graph of the points  
```

```python
# to get the poinst only without joining them can be done using parameter 'o' in plot function 'o' is knowns ring instead of o 
# we can use any argument like *, p....

plt.plot(point_1, point_2, 'o')
plt.show()
```

```python
# multiple line ploting

mul_1 = np.array([0, 1, 2, 3, 4, 5])
mul_2 = np.array([10, 6 , 8, 11, 9, 7])
# here the cosinates will be (0, 10), (1, 6), (2, 8), (3, 11), (4, 9), (5, 7)
```

```python
plt.plot(mul_1, mul_2)
plt.show()
```

```python
plt.plot(mul_1, mul_2, '*')
plt.show()
```

```python
# if we donot specify the x values i.e. if we donot pass the two parameters in plot by default it is considered y-axis values
# then the sefault value of x-axis in matplotlib are uses 0,1,2,3,4,5.... depends on the length of the elements
test = np.array([11, 44, 55, 22, 77, 33, 66])
plt.plot(test)
plt.show()
```

## Markers

* You can use the keyword argument marker to emphasize each point with a specified marker

* marker is basicaly used to show the points of contact through which line passes

* You can use also use the shortcut string notation parameter to specify the marker.

* This parameter is also called fmt, and is written with this syntax:

* marker|line|color

```python
test = np.array([11, 44, 55, 22, 77, 33, 66])
plt.plot(test, marker='*')
plt.show()
```

```python
test_1 = np.array([11, 44, 55, 22, 77, 33, 66])
plt.plot(test_1, 'd--r') # here 2 represents the marker, -- represents type of line and r represents color of the line
plt.show()
```

```python
# we can use ms paremeter to increase the size of the markers
test = np.array([11, 44, 55, 22, 77, 33, 66])
plt.plot(test, marker='p', ms=20)
plt.show()
```

```python
# we can use mec to color the edges of a marker

test = np.array([11, 44, 55, 22, 77, 33, 66])
plt.plot(test, marker='o', mec='r', ms=30)
plt.show()
```

```python
# with mfc parameter the color of the marker can be changed

test = np.array([11, 44, 55, 22, 77, 33, 66])
plt.plot(test, marker='H', ms = 20, mfc='m')
plt.show()
```

```python
test = np.array([11, 44, 55, 22, 77, 33, 66])
plt.plot(test, marker='v', mfc='y', mec='y', ms=20, color='k')# color changes the color of the line
plt.show()
```
