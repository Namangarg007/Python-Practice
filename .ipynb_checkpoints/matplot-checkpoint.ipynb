# Matplotlib

* most of the work in matplot library is done using the modeule pyplot with contains all the functions required

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Plot function

* The plot() function is used to draw points (markers) in a diagram.

* By default, the plot() function draws a line from point to point.

* The function takes parameters for specifying points in the diagram.

* Parameter 1 is an array containing the points on the x-axis.

* Parameter 2 is an array containing the points on the y-axis.



```python
point_1 = np.array([5, 10])
point_2 = np.array([5, 10]) # first parameter in array represents x-axis and second parameter represents y-axis
# the points on axis are (5, 5) and (10, 10)
# we created two 1-D array to represent them in the graph using matplotlib
```

```python
point_1
```

```python
plt.plot(point_1, point_2) # in this point_1 will act as a point on the graph and point_2 will act as another point
plt.show()# this is used to show/display the graph of the points  
```

```python
# to get the poinst only without joining them can be done using parameter 'o' in plot function 'o' is knowns ring instead of o 
# we can use any argument like *, p....

plt.plot(point_1, point_2, 'o')
plt.show()
```

```python
# multiple line ploting

mul_1 = np.array([0, 1, 2, 3, 4, 5])
mul_2 = np.array([10, 6 , 8, 11, 9, 7])
# here the cosinates will be (0, 10), (1, 6), (2, 8), (3, 11), (4, 9), (5, 7)
```

```python
plt.plot(mul_1, mul_2)
plt.show()
```

```python
plt.plot(mul_1, mul_2, '*')
plt.show()
```

```python
# if we donot specify the x values i.e. if we donot pass the two parameters in plot by default it is considered y-axis values
# then the sefault value of x-axis in matplotlib are uses 0,1,2,3,4,5.... depends on the length of the elements
test = np.array([11, 44, 55, 22, 77, 33, 66])
plt.plot(test)
plt.show()
```

## Markers

* You can use the keyword argument marker to emphasize each point with a specified marker

* marker is basicaly used to show the points of contact through which line passes

* You can use also use the shortcut string notation parameter to specify the marker.

* This parameter is also called fmt, and is written with this syntax:

* marker|line|color

```python
test = np.array([11, 44, 55, 22, 77, 33, 66])
plt.plot(test, marker='*')
plt.show()
```

```python
test_1 = np.array([11, 44, 55, 22, 77, 33, 66])
plt.plot(test_1, 'd--r') # here 2 represents the marker, -- represents type of line and r represents color of the line
plt.show()
```

```python
# we can use ms paremeter to increase the size of the markers
test = np.array([11, 44, 55, 22, 77, 33, 66])
plt.plot(test, marker='p', ms=20)
plt.show()
```

```python
# we can use mec to color the edges of a marker

test = np.array([11, 44, 55, 22, 77, 33, 66])
plt.plot(test, marker='o', mec='r', ms=30)
plt.show()
```

```python
# with mfc parameter the color of the marker can be changed

test = np.array([11, 44, 55, 22, 77, 33, 66])
plt.plot(test, marker='H', ms = 20, mfc='m')
plt.show()
```

```python
test = np.array([11, 44, 55, 22, 77, 33, 66, 99, 100, 88])
plt.plot(test, marker='v', mfc='y', mec='y', ms=20, color='k')# color changes the color of the line
plt.show()
```

## line style

* we use linestyle argument during plotting to style the lines in differt ways

```python
p = np.linspace(10, 100, 10, dtype=int)# line space will give an array of 10 equaly spaced value from 
# 10 to 100 according to third argument the first value will be 10 and the last value will be 100 
# and rest 8 values will be equally spaced values between them
```

```python
p
```

```python
plt.plot(p, linestyle = 'dotted')# line style changes the value of the line in the
# form specified now instead of a line we will recieve a dotted line
plt.show()
```

```python
# as shown above the line is dotted insted of straighe line
```

```python
#another argument that changes the shape of the line is ls 
plt.plot(p, ls='dashed')
plt.show()
```

```python
# with color argument we can change the color of the line
plt.plot(p, color='k', ls='dotted') # only predefined color or line style can be specified in the arguments
plt.show()
```

```python
# with linewidth argument the size or thickness of the line can be adjusted
plt.plot(p, linewidth=20)
plt.show()
```

```python
# to print multiple lines on the screen the we have to use two plot method 
# with two different values
plt.plot(p)
plt.plot(test)
plt.show()
```

```python
# the orange line represents test array and the blue line represents p array 
```

### SubPlot

* this function makes two different plots in the same space

* the firs parameter in the subplot function is for number of rows 

* second parameter is for number of columns

* third parameter represents the current plot

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) # first pareneter represents 1 row second represents 2 columns and third represents the 
# below plot this is same for plot 2 
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
# two plots are printed seperately in the same representation
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) # first pareneter represents 1 row second represents 2 columns and third represents the 
# below plot this is same for plot 2 
plt.plot(x,y)
plt.title('Sales')# this will represent the above plot

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title('Share')

plt.suptitle('Companey') # this is super title this will represent the whole graph

plt.show()
```

### scatter plot 

* this is ploted using plt.scater function 

* it takes two arguments and displays only the points instead of lines

```python
plt.scatter(p, test)# p acts as x corrdinates and test acts y corrdinates
plt.show()
```

```python
plt.scatter(a, p)
plt.scatter(a, test)

plt.show()
```

```python
# the color of the points is by default orange and blue but the color can be changed using c argument
plt.scatter(a, p, c='r')
plt.scatter(a, test, c='k')

plt.show()
```

```python
# by pasing list of colors in c argument the points are given different colors
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown"])
plt.scatter(a, test, c=colors)

plt.show()
```

```python
# the color of the points can be changed using colormap which is a predefined library
# in python which has multiple color scales in which each scale has color with values 
# from 0 to 100 we can us color map using cmap argument

plt.scatter(a, test, c=a, cmap='viridis')# viridis is on off the many color map in python
plt.colorbar()# this will display a color bar next to graph of the specified color map
plt.show()
```

```python
# s argument defines the size of the points in the plot
# aplha argument represents the transparancy of the points
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

### Bar plot

* this is ploted using plt.bar function 

* in which the first value represents x-axis and second represents y-axis

* to make horizontal bargraph plt.barh() is used

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.5, color = "hotpink")# width represents the thickness of the bars in the graph
# color represents color of the bars
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```

### Histogram

* A histogram is a graph showing frequency distributions.

* It is a graph showing the number of observations within each given interval.

```python
x = np.random.normal(170, 10, 250)

plt.hist(x, color='r')
plt.show() 
```

### pip plot

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

```python
# the size of the area of each value is determined by mean of each value
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90) # by using start tangent we can change from where the start
# of the pychart will begin by default it is 0 degree or x-axis

plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow=True) # explode seperate the the area from one another 
# by default it id none but by passing a array in this parameter and specifing value for each piece
# the pieces will be seperated from one another
# shadow give shagow to the pie chart by default it is = false
# To add a list of explanation for each wedge, use the legend() function
plt.legend(title='Four Fruits')# title adds a heading to the legend
plt.show() 
```
